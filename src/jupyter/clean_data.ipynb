```python
import pandas as pd
import numpy as np
import re
```

```python
local_abs = '/home/erick/Documents/card_game/src/csv/abs.csv'
df = pd.read_csv(local_abs, sep='|')
df
```

```python
df_grupo = df.groupby('card')['arquetype'].apply(lambda x: ' | '.join(x)).reset_index()
df_grupo.columns = ['card_name', 'arquetype']
df_grupo
```

```python
df_grupo.to_excel('arquetipos_names.xlsx', index=False)
```

```python
monster = '/home/erick/Documents/card_game/src/cache/monster.csv'
spell = '/home/erick/Documents/card_game/src/cache/spell.csv'
trap = '/home/erick/Documents/card_game/src/cache/trap.csv'

df_monster = pd.read_csv(monster, sep=';')
df_spell = pd.read_csv(spell, sep=';')
df_trap = pd.read_csv(trap, sep=';')
```

```python
col_uteis = ['cod', 'card_name', 'tipo', 'especie', 'atk', 'def', 'tcg', 'ocg', 'effect']
df_monster = df_monster[col_uteis].copy()
df_monster
```

```python
col_uteis = ['cod', 'card_name', 'tipo', 'tcg', 'ocg', 'effect']
df_spell = df_spell[col_uteis].copy()
df_spell
```

```python
df_trap = df_trap[col_uteis].copy()
df_trap
```

```python
monster_arq = pd.merge(df_monster, df_grupo, on='card_name', how='left')
monster_arq
```

```python
monster_arq['cod'] = monster_arq['cod'].astype(str)
monster_arq['arquetype'] = monster_arq['arquetype'].fillna('generic')
monster_arq
```

```python
monster_arq.info()
```

# spell

```python
spell_arq = pd.merge(df_spell, df_grupo, on='card_name', how='left')
spell_arq
```

```python
spell_arq['cod'] = spell_arq['cod'].astype(str)
spell_arq['arquetype'] = spell_arq['arquetype'].fillna('generic')
spell_arq
```

```python
spell_arq.info()
```

# Trap

```python
trap_arq = pd.merge(df_trap, df_grupo, on='card_name', how='left')
trap_arq
```

```python
trap_arq['cod'] = trap_arq['cod'].astype(str)
trap_arq['arquetype'] = trap_arq['arquetype'].fillna('generic')
trap_arq
```

```python
trap_arq.loc[trap_arq['cod'].isin(['1157683'])]
```

```python
monster_arq['tcg'] = monster_arq['tcg'].astype(str).str.replace('T02:00:00.000Z', '')
monster_arq['tcg'] = monster_arq['tcg'].astype(str).str.replace('T03:00:00.000Z', '')

monster_arq['ocg'] = monster_arq['ocg'].astype(str).str.replace('T02:00:00.000Z', '')
monster_arq['ocg'] = monster_arq['ocg'].astype(str).str.replace('T03:00:00.000Z', '')

spell_arq['tcg'] = spell_arq['tcg'].astype(str).str.replace('T02:00:00.000Z', '')
spell_arq['tcg'] = spell_arq['tcg'].astype(str).str.replace('T03:00:00.000Z', '')

spell_arq['ocg'] = spell_arq['ocg'].astype(str).str.replace('T02:00:00.000Z', '')
spell_arq['ocg'] = spell_arq['ocg'].astype(str).str.replace('T03:00:00.000Z', '')

trap_arq['tcg'] = trap_arq['tcg'].astype(str).str.replace('T02:00:00.000Z', '')
trap_arq['tcg'] = trap_arq['tcg'].astype(str).str.replace('T03:00:00.000Z', '')

trap_arq['ocg'] = trap_arq['ocg'].astype(str).str.replace('T02:00:00.000Z', '')
trap_arq['ocg'] = trap_arq['ocg'].astype(str).str.replace('T03:00:00.000Z', '')
```

```python
out = '/home/erick/Documents/card_game/src/jupyter/output'
monster_arq.to_excel(f'{out}/card_game_monster.xlsx', index=False)
spell_arq.to_excel(f'{out}/card_game_spell.xlsx', index=False)
trap_arq.to_excel(f'{out}/card_game_trap.xlsx', index=False)
```
