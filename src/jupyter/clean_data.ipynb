```python
local_arquetipos_names = '/home/zeus/Documents/card_game/src/estatistica/jupyter/arquetipos_names.xlsx'
monster = '/home/zeus/Documents/card_game/src/cache/monster.csv'
spell = '/home/zeus/Documents/card_game/src/cache/spell.csv'
trap = '/home/zeus/Documents/card_game/src/cache/trap.csv'

df_local = pd.read_excel(local_arquetipos_names)
df_monster = pd.read_csv(monster, sep='|')
df_spell = pd.read_csv(spell, sep='|')
df_trap = pd.read_csv(trap, sep='|')
```

```python
monster_arq = pd.merge(df_monster, df_local, on='name_card', how='left')
monster_arq
```

```python
monster_arq.columns = ['cod', 'card_name', 'tipo', 'especie', 'ATK', 'DEF', 'TCG', 'OCG', 'effect', 'link_img', 'arquetype']
monster_arq
```

```python
# trocando datas
compile_ = re.compile(r'\d{4}-\d{2}-\d{2}')
monster_arq['TCG'] = list(map(lambda x: compile_.findall(str(x))[0], monster_arq['TCG']))
monster_arq['OCG'] = list(map(lambda x: compile_.findall(str(x))[0], monster_arq['OCG']))
monster_arq
```

# spell

```python
spell_arq = pd.merge(df_spell, df_local, on='name_card', how='left')
spell_arq
```

```python
spell_arq.columns = ['cod', 'name_card', 'tipo', 'identity', 'TCG', 'OCG', 'effect', 'link_img', 'arquetype']
spell_arq
```

```python
spell_arq['TCG'] = list(map(lambda x: compile_.findall(str(x))[0], spell_arq['TCG']))
spell_arq['OCG'] = list(map(lambda x: compile_.findall(str(x))[0] if compile_.findall(str(x)) else '0000-00-00', spell_arq['OCG']))
spell_arq
```

# Trap

```python
trap_arq = pd.merge(df_spell, df_local, on='name_card', how='left')
trap_arq
```

```python
trap_arq.columns = ['cod', 'name_card', 'tipo', 'identity', 'TCG', 'OCG', 'effect', 'link_img', 'arquetype']
trap_arq
```

```python
trap_arq['TCG'] = list(map(lambda x: compile_.findall(str(x))[0], trap_arq['TCG']))
trap_arq['OCG'] = list(map(lambda x: compile_.findall(str(x))[0] if compile_.findall(str(x)) else '0000-00-00', trap_arq['OCG']))
trap_arq
```

```python
monster_arq.to_excel('card_game_monster.xlsx', index=False)
spell_arq.to_excel('card_game_spell.xlsx', index=False)
trap_arq.to_excel('card_game_trap.xlsx', index=False)
```
